```python
import swmm_api
import pyswmm
from pyswmm import Simulation, Links, Nodes, SystemStats
from swmm_api import read_out_file, read_inp_file
from swmm_api.input_file.sections.others import TimeseriesData
from swmm_api.input_file.section_labels import TIMESERIES
import pandas as pd
import numpy as np
```

```python
for num in range(10):
    
    # Setting different inflow
    inflow =[]
    for t in range(360):
        if np.mod(t,10)==0:
            tem = np.abs(2*(num/10)+np.sin(t*np.pi/360)+np.random.random(1)[0]/10)
            key = '3/12/2024'+'\t'+str(0+t//60).zfill(2)+':'+str(t%60).zfill(2)
            inflow.append((key,tem))
    
    inp = read_inp_file('./SWMM/demo_inflow.inp')
    inp[TIMESERIES]['Inflow'] = TimeseriesData('Inflow',inflow)
    inp.write_file('./SWMM/demo_inflow'+str(num)+'.inp')

    
```

```python
# Simulation
sim = Simulation('./SWMM/demo.inp')
sim.start()
sim._model.swmm_end()
sim._model.swmm_close()
```

```python
for num in range(10):
    # 获取out中数据
    out = read_out_file('./SWMM/demo_inflow'+str(num)+'.out')
    df = out.to_frame()

    linkdata = {}
    for i in [1,2,3,4,5]:
        linkdata[str(i)] = {'link_flow':[],'link_volume':[],'link_depth':[]}
        linkdata[str(i)]['link_flow'] += df[('link',str(i),'flow')].iloc[:].values.reshape((-1,1)).tolist()
        linkdata[str(i)]['link_volume'] += df[('link',str(i),'volume')].iloc[:].values.reshape((-1,1)).tolist()
        linkdata[str(i)]['link_depth'] += df[('link',str(i),'depth')].iloc[:].values.reshape((-1,1)).tolist()

    nodedata={}
    for i in [1,2,3,4,5,6]:
        nodedata[str(i)] = {'node_depth':[],'node_volume':[]}
        nodedata[str(i)]['node_depth'] += df[('node',str(i),'depth')].iloc[:].values.reshape((-1,1)).tolist()
        nodedata[str(i)]['node_volume'] += df[('node',str(i),'depth')].iloc[:].values.reshape((-1,1)).tolist()

    np.save('./SWMM_DATA/SWMM_linkdata'+str(num)+'.npy', linkdata)
    np.save('./SWMM_DATA/SWMM_nodedata'+str(num)+'.npy', nodedata)
```
