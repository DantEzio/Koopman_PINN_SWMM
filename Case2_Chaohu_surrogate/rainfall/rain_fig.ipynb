```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
raindatae = np.load(os.path.dirname(os.getcwd())+'\\rainfall\\training_raindata.npy').tolist()
#raindata2top = np.load('./training_rainfall/2top.npy').tolist()
#raindata3top = np.load('./training_rainfall/3top.npy').tolist()
rr = np.load(os.path.dirname(os.getcwd())+'\\rainfall\\real_raindata.npy').tolist()
```

```python
designrain={}
realrain = {}
for i in range(4):
    designrain[i]=[]
    realrain[i]=[]
    for j in range(len(raindatae[i])):
        designrain[i].append(float(raindatae[i][j][1]))
    for j in range(len(rr[i])):
        realrain[i].append(float(rr[i][j][1]))
```

# 选出部分training data和rr data作图

```python
fontx = {
        'fontsize':28,
        'color':'k',
        'family':'Times New Roman',
}
fonty = {
        'fontsize':35,
        'color':'k',
        'family':'Times New Roman', 
}
legend_font = {
        'size':25,
        'family':'Times New Roman', 
}

fig,ax = plt.subplots(8,1,figsize=(18,30))
for i in range(8):
    if i < 4:
        ax[i].bar(np.arange(len(designrain[i])),designrain[i],label='Test rainfall '+str(i+1))
    else:
        ax[i].bar(np.arange(len(realrain[i-4])),realrain[i-4],label='Test rainfall '+str(i+1))
    ax[i].legend(prop=legend_font)
    ax[i].tick_params(axis='y',labelsize=20)
    ax[i].tick_params(axis='x',labelsize=20)
    if i < 7:
        ax[i].set_xticks([])
    else:
        ax[i].set_xlabel('Time (minutes)',fontdict=fontx)

    if i == 3:
        ax[i].set_ylabel('Rainfall intensity (mm/hour)',fontdict=fonty)

plt.savefig(os.path.dirname(os.getcwd())+'\\Figs\\fig2.png',dpi=200,bbox_inches='tight')
```
